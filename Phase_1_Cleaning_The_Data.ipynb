```python
# import the necessary liblaries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset
df = pd.read_csv("/content/Expresso_churn_dataset.csv")
df.head()
```

```python
# Describe the data
df.info()
```

```python
# Drop unecessary columns
df1 = df.drop(columns=['user_id', 'ORANGE', 'TIGO', 'ZONE1', 'ZONE2', 'MRG', 'TOP_PACK'])
df1.head()
```

```python
# Check for missing values
df1["REGION"].isnull().sum()
```

```python
# Checking the number of 'NaN' values in the REGION column that result to the binary CHURN rate of '1'
df1[df1['REGION'].isnull()]['CHURN'].value_counts()
```

```python
# Create a dataframe where the churn binary value = '1'
df2 = df1[df1['CHURN'] == 1]
df2.head()
```

```python
# Check the unique values in the 'REGULARITY' and 'REGION' columns
df2['REGULARITY'].unique()
```

```python
df2['REGULARITY'].value_counts()
```

```python
# Check for the total number of cutsomera who stopped using the service
churne_customers = df2['CHURN'].value_counts()

print(f'The total number of customers who stopped using the service were: {churne_customers}')
```

```python
df2['REGION'].value_counts()
```

```python
# Checking the number of null values in the region column.
df2['REGION'].isnull().sum()
```

```python
# Dropping the null values in df2
df2.dropna(inplace=True)
df2.head()
```

```python
# drop the NaN values in df1 dataframe.
#df1.dropna(inplace=True)
#df1.head()
```

```python
# Check the general information of df1
df1.info()
```

```python
df1['CHURN'].value_counts()
```

```python
# Fill the NaN values in the 'REGION' column with 'NO-REGION'
df1['REGION'].fillna('NO-REGION', inplace=True)
df1.head()
```

```python
# Fill all NaN Values in df1 with o
df1.fillna(0, inplace=True)
df1.head()
```

```python
# Check for duplicates in df1
print(df1.duplicated().sum())

# View the content of the duplicated rows
df1[df1.duplicated()]

# Drop the duplicated rows
df1.drop_duplicates(inplace=True)
print(df1.duplicated().sum())
```
